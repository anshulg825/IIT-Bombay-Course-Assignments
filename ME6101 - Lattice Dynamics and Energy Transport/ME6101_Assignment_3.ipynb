```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math as ma

T=np.linspace(1,600,1200)
df=pd.read_excel('/content/drive/My Drive/Al_DOS_7K.xlsx')
E=df['Energy']
DOS=df['DOS']

def kbt(T):
    return (0.0862*T)
FET=np.zeros(1200)
FE=np.zeros(55)
def n(E,T):
    return (1/((ma.exp(E/T))-1))
def nn(E,T):
    return (1-(ma.exp(-E/T)))
for j in range(1200):
    for i in range (55):
        FE[i]=((E[i]+E[i+1])/4)*((DOS[i]+DOS[i+1])/2)*(E[i+1]-E[i])+(0.0862*T[j])*(ma.log(nn((E[i]+E[i+1])/2,kbt(T[j]))))*((DOS[i]+DOS[i+1])/2)*(E[i+1]-E[i])
    FET[j]=sum(FE)
plt.plot(T,(FET-FET[0])*3/1000)
plt.show()
```

```
ENT=np.zeros(1200)
EN=np.zeros(55)
for j in range(1200):
    for i in range (55):
        EN[i]=(-(0.0862)*(ma.log(nn((E[i]+E[i+1])/2,kbt(T[j]))))+((((E[i]+E[i+1])/2)*(n((E[i]+E[i+1])/2,kbt(T[j]))))/T[j]))*((DOS[i]+DOS[i+1])/2)*(E[i+1]-E[i])
    ENT[j]=sum(EN)
plt.plot(T,ENT*100/3)
plt.show()
```

```
IET=np.zeros(1200)
IE=np.zeros(55)
for j in range(1200):
    for i in range (55):
        IE[i]=(((E[i]+E[i+1])/2)*((1/2)+n((E[i]+E[i+1])/2,kbt(T[j])))*((DOS[i]+DOS[i+1])/2)*(E[i+1]-E[i]))
    IET[j]=sum(IE)
plt.plot(T,(IET-IET[0])*3/1000)
plt.show()
```
